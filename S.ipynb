# Parametric Equations

## Graphing Devices

Most graphing calculators and computer graphing programs can be used to graph curves defined by parametric equations. In fact, it's instructive to watch a parametric curve being drawn by a graphing calculator because the points are plotted in order as the corresponding parameter values increase.

## Example 1
Use a graphing device to graph the curve $x = y^4-3y^2$.

## Solution
If we let the parameter be $t=y$, then we have the equations $x= t^4 - 3t^2$, and $y = t$.
Using these parametric equations we can graph the curve using the following Python code. Note: It would be possible to solve the given equation ($x=y^4-3y^2$) for $y$ as four functions of $x$ and graph them individually, but the parametric equations provide a much easier method. $\square$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=100
t=np.linspace(-3, 3, n+1)
x=t**4-3*t**2
y=t
plt.axis("equal")
plt.grid()
plt.plot(x, y)
plt.show()
```

## Example 2

```python
n=1000
t=np.linspace(0, 2*np.pi, n+1)
x=1.5*np.cos(t) - np.cos(30*t)
y=1.5*np.sin(t) - np.sin(30*t)
plt.axis("equal")
plt.grid()
plt.plot(x, y)
plt.show()
```
